### Figure 1C. pan-genome functional comparison

```python
# load pan-genome function annotation file,
na1011_function = pd.read_excel(r"data/genome/na_pan1011_functional_annotations.xlsx",index_col=0,header=2)
lgpan_function = pd.read_excel(r"data/genome/lg_pan1392_functional_annotations.xlsx",index_col=0,header=2)
this1800_function = pd.read_excel(r"data/genome/pan1800_functional_annotations.xlsx",index_col=0,header=2)

def get_cog_counts(df_annotation,df_cog):
    '''get COG number counts'''
    df_annotation = df_annotation.dropna(subset=["COG_category"])
    cog_list = df_cog.index.tolist()
    cog_counts=pd.Series(index=cog_list)
    for i in cog_list:
        count = len(df_annotation[df_annotation["COG_category"].str.contains(i)])
        cog_counts[i] = count
    return cog_counts

# load COG category,set columns as "COG","category"
df_cog=pd.read_csv("data/cog_cotegory",index_col=0,sep="\t",header=None)
df_cog.columns=["color","description"]
# color列每个元素前加上‘#’
df_cog["color"]='#'+df_cog["color"].astype(str)
lg_cog_counts=get_cog_counts(lgpan_function,df_cog)
na_cog_counts=get_cog_counts(na1011_function,df_cog)
this_cog_counts=get_cog_counts(this1800_function,df_cog)
df_cog_counts=pd.DataFrame(columns=['Sce-pan1011','Sce-pan1364',"Sce-pan1807"])
df_cog_counts["Sce-pan1364"]=lg_cog_counts
df_cog_counts["Sce-pan1011"]=na_cog_counts
df_cog_counts["Sce-pan1807"]=this_cog_counts
# remove cog that all counts are less than 10
df_cog_counts=df_cog_counts[(df_cog_counts>10).any(axis=1)]
df_cog_counts=df_cog_counts.astype(int)

# reset order of rows
df_cog_counts=df_cog_counts.reindex(index=df_cog_counts.index[::-1])
# reset index of df_cog according to df_cog_counts
df_cog=df_cog.reindex(index=df_cog_counts.index)

# change column order as na1011_pangenome,lg1362_pangenome,this1800_pangenome
df_cog_counts=df_cog_counts[['Sce-pan1011','Sce-pan1364',"Sce-pan1807"]]

cog_list=df_cog_counts.index.tolist()
# sort according to the Sce-pan1011 value
df_cog_counts=df_cog_counts.sort_values(by='Sce-pan1011',ascending=False)
df_cog_counts
```

```python
# plot grouped barplot to compare COG gene count in each COG category for 3 pangenome
fig,ax=plt.subplots(figsize=(2.5,6.7),dpi=400)
# set figure style
# plt.style.use('seaborn-paper')
# set figure background color
plt.rcParams['figure.facecolor'] = 'white'
# set Dark2 color palette
colors_set=sns.color_palette('Dark2')
bar_width = 0.25
index = np.arange(len(cog_list))
# plot bar plot in ax, orientation='horizontal'
ax.barh(y=index,
        width=df_cog_counts['Sce-pan1011'],
        height=bar_width,
        label='Sce-pan1011 (%s)'%len(na1011_function),
        color=colors_set[0])
ax.barh(y=index-bar_width,
        width=df_cog_counts['Sce-pan1364'],
        height=bar_width,
        label='Sce-pan1364 (%s)'%len(lgpan_function),
        color=colors_set[1])
ax.barh(y=index-bar_width*2,
        width=df_cog_counts['Sce-pan1807'],
        height=bar_width,
        label='Sce-pan1807 (%s)'%len(this1800_function),
        color=colors_set[2])

ax.set_yticks(index-bar_width)
cog_description=df_cog[df_cog.index.isin(cog_list)]["description"].tolist()
ax.set_yticklabels(cog_description,fontsize=12)
ax.set_xlabel('Gene number',fontsize=12,fontdict=axis_fontdict)
ax.set_ylabel('COG categories',fontsize=12,fontdict=axis_fontdict)
# ax.set_title('Fuctional comparison of 3 pan-genomes',fontdict=title_fontdict)
ax.legend(fontsize=10)
plt.show()
# save figure
fig.savefig(r'figures/output/figure1_pan-genome_functional_comparison.svg',bbox_inches='tight',dpi=400,transparent=True)
```
