### venn plot show the essential rxns/genes in aerobic/anaerobic condition

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir(r'D:\code\github\Unified_Yeast_GEMs_Database_from_13pro\Unified_Yeast_GEMs_Database')
```

```python

# load simulation results
df_aerobic_anaerobic_essential_geneList=pd.read_excel('result/ssGEM_simulation/panYeast_anaerobic&aerobic_related_genes.xlsx',index_col=0)
df_aerobic_anaerobic_essential_rxnList=pd.read_excel('result/ssGEM_simulation/panYeast_anaerobic&aerobic_essential_rxns.xlsx',index_col=0)

anaerobic_geneList=df_aerobic_anaerobic_essential_geneList[df_aerobic_anaerobic_essential_geneList['anaerobic_related']==1].index.tolist()
aerobic_geneList=df_aerobic_anaerobic_essential_geneList[df_aerobic_anaerobic_essential_geneList['aerobic_related']==1].index.tolist()

anaerobic_rxnList=df_aerobic_anaerobic_essential_rxnList[df_aerobic_anaerobic_essential_rxnList['anaerobic_essential']==1].index.tolist()
aerobic_rxnList=df_aerobic_anaerobic_essential_rxnList[df_aerobic_anaerobic_essential_rxnList['aerobic_essential']==1].index.tolist()
```

```python
# plot venn diagram for aerobic and anaerobic essential genes
from matplotlib_venn import venn2
plt.figure(figsize=(4,6))
venn2([set(anaerobic_geneList),set(aerobic_geneList)],set_labels=['anaerobic condition related genes','aerobic condition related genes'])
plt.title('Venn diagram for aerobic and anaerobic essential genes',fontsize=15,weight='bold')
plt.show()
```

```python
# plot venn diagram for aerobic and anaerobic essential rxns
from matplotlib_venn import venn2
plt.figure(figsize=(4,6))
venn2([set(anaerobic_rxnList),set(aerobic_rxnList)],set_labels=['anaerobic condition related rxns','aerobic condition related rxns'])
plt.title('Venn diagram for aerobic and anaerobic essential rxns',fontsize=15,weight='bold')
plt.show()
```
