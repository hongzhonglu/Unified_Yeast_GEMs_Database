```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#set working dir
import os
os.chdir(r"D:\code\github\Unified_Yeast_GEMs_Database_from_13pro\Unified_Yeast_GEMs_Database")
```

```python
#load data
df_busco=pd.read_csv("code/3.pan-genome_construction/3.pan-genome_comparison/output/pangenomes_busco_scores.txt",index_col=0)
df_busco
```

```python
df_busco.iloc[:,1:5]
```

#### 1.compare the 3 pangenome BUSCO result

```python
df_pangenome_busco=df_busco.reindex(index=['lgpangenome','na1011_pangenome','pan1800_v2'])
df_pangenome_busco
```

```python
# plot the barh plot 
sns.set()
# plot the barh plot
df_pangenome_busco.iloc[:,1:5].plot.barh(stacked=True,figsize=(10,5))
# set the legend outside the plot
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.title("BUSCO scores of 3 pangenomes",fontsize=15,weight='bold')
# set x axis label
plt.xlabel("BUSCO scores(%)",fontsize=12,weight='bold')
# set y axis label
plt.ylabel("Pangenome",fontsize=12,weight='bold')
# rotate the y axis label
plt.yticks(rotation=30)
plt.show()
```

```python
# plot the group bar plot to comare the S,D,F,M scores of 3 pangenomes
sns.set()
# plot the barh plot
df_pangenome_busco.iloc[:,:5].T.plot.bar(figsize=(10,5))
# set the legend outside the plot,and reset the legend label
plt.legend(bbox_to_anchor=(1.0, 1.0),title='Pangenome',labels=['lg1392 pangenome','na1011 pangenome','this study'])
plt.title("BUSCO evaluation of 3 pangenomes",fontsize=15,weight='bold')
# set x axis label
plt.ylabel("BUSCO scores(%)",fontsize=12,weight='bold')
# set y axis label
plt.xlabel("BUSCO category",fontsize=12,weight='bold')
# reset x ticks name ,rotate the x axis label
plt.xticks(ticks=range(5),labels=['Total complete','Single-copy complete','Duplicated complete','Fragmented','Missing'],rotation=30)
plt.show()
```
