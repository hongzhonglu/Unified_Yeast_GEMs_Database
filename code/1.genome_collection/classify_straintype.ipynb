redefine the S.cerevisiae strain type according to strain sub type

```python
import pandas as pd
import os

# set work dir
os.chdir(r'D:\code\github\Unified_Yeast_GEMs_Database_from_13pro\Unified_Yeast_GEMs_Database')
```

```python
all_strain_info=pd.read_excel(r"data\1897_strains_info.xlsx",index_col=0)
all_strain_info['sub_type'].value_counts()
```

- redefine strain type:<br>

|type|sub types|
|-|-|
|Wild|Wild,Insect,Soil|
|Fermentation|Fermentation,Dairy,Bakery|
|Human|Human|
|Wine|Sake,Wine,Beer,Distillery,Cider|
|Lab|Lab|
|Industry|Bioethanol,Industrial|
|Unknow|0|

```python
# reset strain type
all_strain_info2=all_strain_info.copy()
all_strain_info2['type']=None
# wild
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Wild","Insect","Soil"]),'type']='Wild'
# Fermentation
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Fermentation","Dairy","Bakery"]),'type']='Fermentation'
# Human
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Human"]),'type']='Human'
# Wine
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Wine","Sake","Beer","Distillery","Cider"]),'type']='Wine'
# Lab
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Lab"]),'type']='Lab'
# Industry
all_strain_info2.loc[all_strain_info2['sub_type'].isin(["Industrial","Bioethanol"]),'type']='Industry'
# Unknow
all_strain_info2.loc[all_strain_info2['sub_type'].isin([0]),'type']='Unknow'

all_strain_info2['type'].value_counts()
```

```python
# save result
all_strain_info2.to_excel(r"data\1897_strains_info.xlsx")
```
