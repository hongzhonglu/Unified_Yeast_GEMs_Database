```python
import sys
sys.path.append(r"D:\code\github/Unified_Yeast_GEMs_Database\code")
from mainFunction import get_gene_lens

def parse_blastp_result(blastp_file,blastp_dir,query,query_dir,subject,subject_dir):
    """
    parse the blastp result file and calculate COV for query and subject
    :param blastp_file: blastp result file
    :param query: query file name
    :param query_dir: query file directory
    :param subject: subject file name
    :param subject_dir: subject file directory
    :return: blastp_file with COV for query and subject
    """
    # load blastp result file
    df_blastp_file = pd.read_csv(blastp_dir+blastp_file, sep="\t", header=None, index_col=0)
    columns = ["subject", "identity", "alignment length", "mismatches", "gap opens", "q_start", "q_end", "s_start",
               "s_end", "evalue", "bit score"]
    df_blastp_file.columns = columns

    # get gene length for query and subject
    query_lens=get_gene_lens(query,in_folder=query_dir)
    subject_lens=get_gene_lens(subject,in_folder=subject_dir)
    query_lens.set_index("gene",inplace=True)
    subject_lens.set_index("gene",inplace=True)

    # map query lens to blastp_file and name the column as "query_lens"
    df_blastp_file=df_blastp_file.join(query_lens,how="left")
    df_blastp_file.rename(columns={"gene_length":"query_lens"},inplace=True)
    # map subject lens to blastp_file and name the column as "subject_lens"
    df_blastp_file=df_blastp_file.join(subject_lens,how="left",on="subject")
    df_blastp_file.rename(columns={"gene_length":"subject_lens"},inplace=True)

    # calculate COV for query and subject
    df_blastp_file["query_cov"]=(df_blastp_file["q_end"]-df_blastp_file["q_start"]+1)/df_blastp_file["query_lens"]
    df_blastp_file["subject_cov"]=(df_blastp_file["s_end"]-df_blastp_file["s_start"]+1)/df_blastp_file["subject_lens"]

    return df_blastp_file
```

```python
# pan1011 vs pan1011
df_blastp_1011_vs_1011=parse_blastp_result(blastp_dir='code/3.pan-genome_construction/3.pan-genome_comparison/output/',
                                           blastp_file='pan1011_vs_pan1011_blastp.txt',
                                           query='na_pan1011.fasta',
                                           query_dir='data/genome/',
                                           subject='na_pan1011.fasta',
                                           subject_dir='data/genome/')

pan1011_all_geneID=list(set(df_blastp_1011_vs_1011["subject"]))
```

```python
threshold=0.95
df_filter=df_blastp_1011_vs_1011[(df_blastp_1011_vs_1011["query_cov"]>=threshold)&
                                 (df_blastp_1011_vs_1011["subject_cov"]>=threshold)&
                                 (df_blastp_1011_vs_1011["identity"]>=threshold)]
# remove index equal to subject
df_filter=df_filter[df_filter["subject"]!=df_filter.index]
```

```python
# randomly sample 1011 strains from 1807
pan1807_geneMatrix=pd.read_csv(r'data/geneMatrix/pan1800_v2_blastp_50_70_geneMatrix.csv',index_col=0)
# remove all .fa in column
pan1807_geneMatrix.columns=[i.replace('.fa','') for i in pan1807_geneMatrix.columns]

all_strainList=list(pan1807_geneMatrix.columns)

def coregenome_analysis(geneMatrix,strainList,core_threshold=0.99):
    # check if all strains are in geneMatrix
    if set(strainList).issubset(set(geneMatrix.columns)):
        # get the geneMatrix for the strains
        geneMatrix_sub=geneMatrix[strainList]
        # calculate the core genome
        gene_ratio=geneMatrix_sub.apply(lambda x:sum(x)/len(x),axis=1)
        coregenome=gene_ratio[gene_ratio>=core_threshold].index.tolist()
        return coregenome
    else:
        print("Not all strains are in the geneMatrix")

# randomly sample 1011 strains from 1807 100 times
import random
coregenome_list=list()
for i in range(1000):
    strainList=random.sample(all_strainList,1011)
    coregenome=coregenome_analysis(pan1807_geneMatrix,strainList)
    numb=len(coregenome)
    coregenome_list.append(numb)
```
